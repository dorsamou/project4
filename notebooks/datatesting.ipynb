```python
asset = "HopkinsBuilding"
data_set = DATASETS[asset]

if data_set["type"] == "load":
    formatter = BuildingLoadFormatter(
        data_set["filename"],
        cal_real=data_set["cal_real"],
        cal_reactive=data_set["cal_reactive"],
    )
    df = formatter.run(save=False)

elif data_set["type"] == "pv":
    formatter = PVGeneratorFormatter(data_set["filename"])
    df = formatter.run(save=False)

elif data_set["type"] == "battery":
    formatter = BatteryStorageFormatter(
        data_set["filename"],
        rating_kw=data_set["rating_kw"]
    )
    df = formatter.run(save=False)

elif data_set["type"] == "loadwev":
    formatter = BuildingLoadFormatter(
        data_set["filename"],
        cal_real=data_set["cal_real"],
        cal_reactive=data_set["cal_reactive"],
    )
    df = formatter.run(save=False)
    ev_asset = asset.replace("Building", "EV")
    ev_data_set = DATASETS[ev_asset]
    ev_formatter = BuildingLoadFormatter(
        ev_data_set["filename"],
        cal_real=ev_data_set["cal_real"],
        cal_reactive=ev_data_set["cal_reactive"],
    )
    ev_df = ev_formatter.run(save=False)

    df = combine_power(df, ev_df)


df.head()
```
